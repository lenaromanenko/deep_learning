```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.mobilenet import preprocess_input
from google.colab import drive
from tensorflow.keras.applications import MobileNet
from pprint import pprint
from tensorflow.keras.applications.mobilenet import decode_predictions
```

```python
def Image(URL):

  im = image.load_img(URL, target_size=(224, 224))
  a = image.img_to_array(im)
  a = a.reshape(1, 224, 224, 3)
  a = preprocess_input(a)

  drive.mount('/content/drive')

  m = MobileNet(input_shape=(224, 224, 3))
  m.compile(optimizer='rmsprop', loss='categorical_crossentropy',
           metrics=['accuracy'])
  p = m.predict(a)
  pprint(decode_predictions(p, 10))
  print('Image:')
  return im

Image('/content/drive/MyDrive/Colab Notebooks/MobileNet/balloon.jpg')
```
